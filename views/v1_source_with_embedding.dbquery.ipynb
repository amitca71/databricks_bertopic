```sql

create or replace view v1_source_with_embedding as
(
with t_source as (select id, text, translated from arabic_offensive_comment_detection_annotation_input),
t_embedding as (select dataset_name, embedding_model, id ,text_embedding from embeded_text where embedding_status='success')
select t_source.id,t_embedding.embedding_model, t_embedding.dataset_name ,t_source.text, t_source.translated, t_embedding.text_embedding from t_source inner join t_embedding using (id)
)
```
