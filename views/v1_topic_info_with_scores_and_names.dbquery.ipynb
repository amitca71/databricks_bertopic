```sql

create or replace view v1_topic_info_with_scores_and_names as(
with t_topic_info as (select execution_id, father_topic, layer, topic, count from topic_info_local),
t_model_evaluation as (select execution_id, layer,father_topic, topic, json_tuple(ai_grade, 'score', 'description') AS (score, score_description) from model_evaluation),
t_topic_naming as (select execution_id, layer,father_topic, topic,  json_tuple(topic_name, 'title', 'description') AS (title, description) from topic_naming)
select concat(CAST(t_topic_info.father_topic AS STRING), "_", CAST(t_topic_info.topic AS STRING)) as full_topic ,t_topic_info.*, score, score_description, title, description from t_topic_info
inner join t_model_evaluation
on t_topic_info.execution_id = t_model_evaluation.execution_id
and t_topic_info.topic = t_model_evaluation.topic
and t_topic_info.layer = t_model_evaluation.layer
and t_topic_info.father_topic = t_model_evaluation.father_topic
inner join t_topic_naming
on t_topic_info.execution_id = t_topic_naming.execution_id
and t_topic_info.topic = t_topic_naming.topic 
and t_topic_info.layer = t_topic_naming.layer
and t_topic_info.father_topic = t_topic_naming.father_topic)
```
